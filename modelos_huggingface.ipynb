Usaremos pipeline, esta funcion se encarga de todo el preprocesamiento y devuelve predicciones depuradas. El pipeline se utiliza principalmente para la inferencia, donde aplicamos modelos ajustados a nuevos ejemplos.

MÁS INFO
https://huggingface.co/docs/transformers/main/en/main_classes/pipelines

```python
from transformers import pipeline
```

La estructura que usamos para cargar modelos es:

NOMBRE = pipeline(task="TAREA_DEL_MODELO")

Esta estrutura escogera el modelo mas optimo para la tarea.

si quisieramos darle un modelo que nosotros escoguimos seria:

NOMBRE = pipeline(task="TAREA_DEL_MODELO",model="NOMBRE_DEL_MODELO")

Para ontener respuesta usamos la siguiente estructura.

NOMBRE("ENTRADA", atributos)

# Modelos de Texto

```python
resumen = pipeline(task="summarization")
```

```python
texto ="""A partir de las 20:00 del sábado 18 de febrero, los servicios digitales de la banca privada se empezaron a restituir de manera paulatina, luego de varias horas de intermitencias, que generaron malestar entre los clientes al no poder realizar transacciones bancarias durante la tarde del primer día del feriado de carnaval.
El proveedor de este servicio para la banca privada, la empresa Telconet Latam Ecuador, pidió disculpas públicas “por este incidente, único durante once años de operación continua e ininterrumpida”. En el mensaje que fue posteado por el dueño de la empresa, Tomislav Topic, además se indicó que se encontraban trabajando en la resolución total y definitiva del mismo.
Durante la tarde del sábado varios clientes reportaron que no pudieron hacer pagos con sus tarjetas de crédito y débito, tampoco podían obtener efectivo de los cajeros. Las quejas de los usuarios se repetían desde restaurantes, gasolineras, cines y diversos comercios. Los usuarios de varios bancos como de servicios bancarios levantaron las alarmas al respecto. Incluso este domingo, 19 de febrero, a primeras horas de la mañana, todavía se podían registrar quejas de los clientes que indicaban que no estaban totalmente restituidos los servicios.
"""
resumen(texto, min_length = 45)
```

Como vemos, este esta optimizado para ingles y no da buenos resultados en español, por lo que deberiamos escoger uno es español para que el resultado sea mas optimo.



```python
resumen_es = pipeline(
    task="summarization",
    model="IIC/mt5-spanish-mlsum"
)
```

```python
texto = """El ejército ruso afirmó el miércoles que derribó o dejó fuera de servicio diez drones ucranianos en un intento de ataque “masivo” contra instalaciones en la anexionada península de Crimea, tras una incursión similar la víspera con esos artefactos en territorio ruso.
“Un intento de ataque masivo con drones por parte del régimen de Kiev en las instalaciones de la península de Crimea ha sido frustrado”, anunció el Ministerio de Defensa ruso en un comunicado.
Según el ministerio, la defensa antiaérea rusa derribó seis drones de ataque ucraniano y otros cuatro fueron desactivados por medios electrónicos.
El martes, el territorio ruso fue objeto de un ataque de drones, que no dejó daños.
Por primera vez, uno de esos aparatos se estrelló en la región de la capital, Moscú.
Otros tres drones fueron derribados en la noche el martes de madrugada en otras regiones.
No es la primera vez que Crimea, península anexada por Rusia en 2014 donde tiene base su flota del mar Negro y desde donde lleva a cabo parte de su ofensiva en Ucrania, es atacada por drones.
Aparatos sin piloto ya habían apuntado contra el puerto de Sebastopol en diciembre. En octubre, el blanco fue una central eléctrica y en agosto, el Estado Mayor de la flota rusa del mar Negro.
"""

resumen_es(texto, min_length=50, max_length=100)
```

```python
Clasificacion_sentimiento = pipeline(task="text-classification", model="pysentimiento/robertuito-sentiment-analysis")
```

```python
textos = ["Las papas fritas son deliciosas",
         "Que feo es cuando el mercado cae."]

Clasificacion_sentimiento(textos)
```
